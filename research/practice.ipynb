```python
import pandas as pd
import json

# 1. Load the data
df = pd.read_csv('Book1.csv')

# 2. Basic processing (Optional)
# Let's say we want to see the growth rate between months
df['Growth'] = df['Actual_Revenue'].pct_change().fillna(0) * 100

# 3. THE MAGIC STEP: Convert to Recharts Format
# Recharts needs a list of objects: [{"Month": "Jan", "Actual_Revenue": 4000}, ...]
chart_ready_data = df.to_dict(orient='records')

# 4. Save it as a JSON file or just print it for the frontend
print("--- COPY AND PASTE THIS INTO YOUR REACT APP ---")
print(json.dumps(chart_ready_data, indent=2))

# In a real app, your FastAPI route would simply:
# return chart_ready_data
```
